```python
import matplotlib.pyplot as plt # Import matplotlib for graphs and visualizations
%matplotlib inline
import random # Random number generator
import numpy as np # Import numpy for mathematical operations
from scipy.stats import norm

from ipywidgets import widgets # Interactive visualizations
from IPython.html.widgets import *
```

### We can test effect of sample size using interactive IPython visualizations

```python
def plot_uniform_sample(f):
    sample = list()
    for i in range(int(np.power(10,f))):
        sample.append(random.uniform(0,1))
    plt.hist(sample)
    plt.title('Sample size: {}'.format(len(sample)))

interact(plot_uniform_sample, f=(1,5,0.1))
```

```python
def plot_gaussian_sample(f):
    sample = list()
    for i in range(int(np.power(10,f))):
        sample.append(random.gauss(0,1))
    plt.hist(sample)
    plt.title('Sample size: {}'.format(len(sample)))

interact(plot_gaussian_sample, f=(1,5,0.1))
```

### Let's also try how normal distribution effected by it's two parameter: mean and std

```python
def plot_gaussian_dist_mean(f):
    sample = [random.gauss(f,1) for i in range(1000)]
    plt.hist(sample, normed=True, label='Sample (1000)')
    plt.title('Mean:{:.2f}, Std:{}'.format(f, 0))
    
    xrange = np.linspace(-3,3,100)
    plt.plot(xrange, norm.pdf(xrange), 'r', linewidth=2, label='Theory N(0,1)')
    
    plt.legend()
    
interact(plot_gaussian_dist_mean, f=(-5,5,0.1))
```

```python
def plot_gaussian_dist_std(f):
    sample = [np.random.normal(0,f) for i in range(1000)]
    plt.hist(sample, normed=True, label='Sample (1000)')
    plt.title('Mean:{}, Std:{:.2f}'.format(0, f))
    
    xrange = np.linspace(-3,3,100)
    plt.plot(xrange, norm.pdf(xrange), 'r', linewidth=2, label='Theory N(0,1)')
    
    plt.legend()
    
interact(plot_gaussian_dist_std, f=(0,5,0.1))
```
