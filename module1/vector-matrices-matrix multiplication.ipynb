# Vectors

A vector is a mathematical object generally used to represent physical quantities that have both magnitude and direction. A vector is typically denoted placing an arrow on the top a Latin letter, as for example $\vec{x}$. 

A vector is composed of an arbitrary number of components, each referring to a specific coordinate. For example, the vector

$
\vec{x} = 
\left(
\begin{array}{c}
x_1
\\
x_2
\end{array}
\right)
$

contains two coordinates that specify the position of 
a point in a $2$-dimensional space (i.e., the plane). 
In graphical visualizations, vectors are generally represented
as directed segments, starting from the origin and pointing
to the coordinates specified by the vector components.
In the following, we provide three examples of 
vectors with two components, thus living in the plane.

```python
##define and visualize three vectors in the plane

import matplotlib.pyplot as plt
%matplotlib inline  


##first vector
a = [2, 3]

##second vector
b = [-2, 5]

##third vector
c = [-2, -4]


plt.figure(figsize=(10,10))
plt.rc('text', usetex=True)
plt.rc('font', size=24, **{'family':'sans-serif','sans-serif':['Helvetica']})
plt.rcParams['xtick.major.pad'] = 8
plt.rcParams['ytick.major.pad'] = 8

plt.xlim(-3, 3)
plt.ylim(-5, 6)

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')


plt.arrow( 0.0, 0.0, a[0], a[1], fc='black', ec='black', head_width=.05, head_length=.1, linewidth=3)
plt.arrow( 0.0, 0.0, b[0], b[1], fc='red', ec='red', head_width=.05, head_length=.1, linewidth=3)
plt.arrow( 0.0, 0.0, c[0], c[1], fc='blue', ec='blue', head_width=.05, head_length=.1, linewidth=3)

plt.show()
```

A vector defined in a $N$-dimensional space has $N$ components

$
\vec{x} = 
\left(
\begin{array}{c}
x_1
\\
x_2
\\
\vdots
\\
x_N
\end{array}
\right)
$

Vectors can be represented as column vectors, as in the example reported above, or as
row vectors. In the latter case, the components of the vector are listed horizontally

$
\vec{x} = 
\left(
\begin{array}{c c c c}
x_1,
&
x_2,
&
\ldots,
&
x_N
\end{array}
\right) .
$

An element in a vector is specified by a single interger index.
For example, in the vector with $5$ components

$
\vec{x} = 
\left(
\begin{array}{c c c c c}
100,
&
-3,
&
\pi,
&
10^3
&
1
\end{array}
\right)
$

the third component of the vector is $x_3 = \pi$, while the
first component of the vector is $x_1 = 100$.

## Vectors in python

Vectors are frequently used in computer programs, although they may be named differently
, e.g. array, list. For example, a list in python is a vectorial object. Entries of a list can be accessed by specifying their position. Note that the first entry corresponds to the index zero. Entries of list, however, do not contain necessarily numbers, but they can contain more complicated objects, as for example strings.



# Matrices

A matrix is a rectangular array of numbers, symbols, or expressions, arranged in rows and columns. They are usually represented as a sort of tables. For example, the following is a
$2 \times 3$ matrix

$
A = \left(
\begin{array}{c c c}
3 & \pi & 10^5
\\
-900 & 6.5 & 1
\end{array}
\right) .
$

A matrix is said to be a $N \times M$ matrix if it has $N$ rows and $M$ columns.
The total number of elements or entries of a matrix is given by the product of the
number of rows times the number of columns. For example, the matrix $A$ reported above
has $6$ entries. If the number of rows is equal to the
number of columns, the matrix is a square matrix.


Every entry of a matrix is fully specified by two indices. The first usually refers to the row, while the second to the index of the column. In the matrix reported
above for example, we have $A_{1, 2} = \pi$, $A_{2, 3} = 1$, and $A_{1, 1}=3$.

Note that a vector is a matrix. A column vector is a matrix with a single column. A row vector is instead a matrix with a single row. 


## Examples of matrices

A matrix is a very convenient way to store information about relations
between pairs of objects. We saw already various examples of matrices 
during the lectures, as for example transition matrices in Markov chains and adjacency matrices of networks. We report below some additional details on these two types of matrices.

In a transition matrix, the element $(i, j)$ of the matrix represents the probability to pass from state $i$ to state $j$ at the next stage of the stochastic process. For example, if we model a fair coin toss with a Markov chain model, where every stage consists in tossing the coin, and we indicate with state $1$ head and with state $2$ tail, the transition matrix is given by

$
T = \left(
\begin{array}{cc}
0.5 & 0.5
\\
0.5 & 0.5
\end{array}
\right) .
$

The probability to obtain head, after we had obtained head at previous stage is $T_{1, 1} = 0.5$. The same is also true for any other possibility, being the coin fair, and being every stage independent on the outcome of the previous stage.


The adjacency matrix $A$ of an undirected and unweighted graph is a square matrix, where the element
$(i, j)$ of the matrix is equal to one if there is a connection between nodes $i$ and $j$, while is equal to zero, otherwise. For example the adjacency matrix

$
A = \left(
\begin{array}{ccc}
0 & 1 & 1
\\
1 & 0 & 0
\\
1 & 0 & 0
\end{array}
\right) 
$

represents a graph composed of three nodes, where node $1$ is connected
to both vertices $2$ and $3$. Vertices $2$ and $3$ are however not directly connected. The adjacency matrix of an undirected graph is always symmetric, in thre sense that $A_{i,j} = A_{j, i}$, for all $(i, j)$.

## Matrices in python

Matrices are frequently used in computer programs, although they may be named differently, e.g. arrays of arrays, lists of lists. The library [numpy](http://www.numpy.org) provides a easy way to deal with matrices in python. In the following, we show a simple example.



```python
import numpy as np


##generates a matrix composed of 5 rows and 2 columns
##entries are set all equal to zero
matrix = np.zeros((5, 2))


##we can access every element of the matrix
##specifying two indices. Remember that
##the first index is always zero

##print element in the first row, first column
print (matrix[0][0])


##set element in the first row, first column
##equal to 100

matrix[0][0] = 100
print (matrix[0][0])




##print element in the fifth row, second column
print (matrix[4][1])


##set element in the fifth row, second column
##equal to -1

matrix[4][1] = -1
print (matrix[4][1])

```

# Matrix multiplication

## Matrix multiplication by a scalar

Any $N \times M$ matrix (remember that vectors are also matrices) multiplied by a number (or scalar) is equal to a matrix
$N \times M$ where every entry of the original matrix is
simply multiplied by the scalar. For example


$
5 \cdot \; \left(
\begin{array}{ccc}
0 & 1 & 2
\\
3 & 0 & 4
\\
1 & 5 & 0
\end{array}
\right) 
=
\left(
\begin{array}{ccc}
0 & 5 & 10
\\
15 & 0 & 20
\\
5 & 25 & 0
\end{array}
\right) .
$

In the example above, the number $5$ is the scalar that
multiplies the matrix.



## Row-column product among matrices

Here, we illustrate the definition of the so-called row-column product among matrices. Let us indicate the operation as

$
A \cdot B = C
$

in the sense that the matrices $A$ and $B$ are multiplied, and the
result of this operation is a new matrix $C$. This operation is defined only if it involves a $N \times M$ matrix $A$ and a  $M \times L$ matrix $B$, in the sense that the number of columns of first matrix must be equal to the number of rows of the second matrix. The result of this operation is a $N \times L$ matrix $C$. The rules of the product are very simple. The element $(i, j)$ of the matrix $C$ is computed as

$
C_{i, j} = \sum_{m=1}^M  \; A_{i,m} \, B_{m,j} .
$

This means that we fix the row index $i$ of matrix $A$ and the column index $j$
of matrix $B$, and we make the sum of the products of
all corresponding entries along the row in $A$ and the column in $B$
to obtain the element $C_{i,j}$. We can repeat the same exact operation
for all possible choices of $i$ and $j$ to construct the entire matrix $C$. A simple example is provided below

$
\left(
\begin{array}{cc}
1 & 3
\\
3 & -1
\end{array}
\right)
\cdot
\left(
\begin{array}{ccc}
0 & 1 & 1
\\
2 & -1 & 2
\end{array}
\right)
=
\left(
\begin{array}{ccc}
1 \cdot 0 + 3 \cdot 2  & 1 \cdot 1 + 3 \cdot (-1) & 1 \cdot 1 + 3 \cdot 2 
\\
3 \cdot 0 + (-1) \cdot 2 & 3 \cdot 1 + (-1) \cdot (-1) & 3 \cdot 1 + (-1) \cdot 2
\end{array}
\right)
=
\left(
\begin{array}{ccc}
6  & -2 & 7
\\
-2 & 4 & 1
\end{array}
\right).
$


Below, we report the python implementation of this example.

```python
import numpy as np

##########################
def matrix_multiplication (A, B):
    ##check dimensions
    if A.shape[1] != B.shape[0]:
        print ('Error: matrices must have proper dimensions\n')
        return
        
    N = A.shape[0]
    M = A.shape[1]
    L = B.shape[1]
    
    C = np.zeros((N, L))

    for i in range(0, N):
        for j in range(0, L):
            tmp = 0.0
            for m in range(0, M):
                tmp = tmp + A[i][m] * B[m][j]
            C[i][j] = tmp
            
    return C
#############################


A = np.zeros((2, 2))
A[0][0] = 1.0
A[0][1] = 3.0
A[1][0] = 3.0
A[1][1] = -1.0


B = np.zeros((2, 3))
B[0][0] = 0.0
B[0][1] = 1.0
B[0][2] = 1.0
B[1][0] = 2.0
B[1][1] = -1.0
B[1][2] = 2.0



C = matrix_multiplication (A, B)


print ('A = ', A, '\n')
print ('B = ', B, '\n')
print ('A B = C = ', C, '\n')


##this should return error
C = matrix_multiplication (B, A)




##numpy function
print ('\n\n\nNumpy\n')
C = np.dot(A, B)
print ('A B = C = ', C, '\n')
```
