```python
import matplotlib.pyplot as plt # Import matplotlib for graphs and visualizations
%matplotlib inline

import random # Random number generator

import numpy as np # Import numpy for mathematical operations
```

[Identity matrix](https://en.wikipedia.org/wiki/Identity_matrix): Is a unit matrix with ones on the main diagonal and zeros elsewhere. Multiplication of matrix A and identity matrix equal to A itself.

$
I = \begin{bmatrix}
1 & 0 & \cdots & 0\\ 
0 & 1 & \cdots & 0\\ 
\vdots  & \vdots  & \ddots  & \vdots\\ 
0 & 0 & \cdots & 1
\end{bmatrix}
$

$I_m A = A I_n = A$

```python
# Identity matrix

I = np.eye(2)
print('Identity matrix I=\n{}'.format(I))

a = np.array([2,3])

# b = I * a
b = np.dot(I,a)
print('\nVector b={}'.format(b))
```

Arithmetic operations like summation and substraction can be applied to vectors and matrices with the same sizes.

$
\begin{bmatrix}
a \\ b \\ c
\end{bmatrix}
+
\begin{bmatrix}
x \\ y \\ z
\end{bmatrix}
= 
\begin{bmatrix}
a+x \\ b+y \\ c+z
\end{bmatrix}
$

```python
vecA = np.array([1,3])
vecB = np.array([3,1])
vecC = np.array([-2,-4])

fig = plt.figure(figsize=(5,5))
plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.xlabel('$x_1$')
plt.ylabel('$x_2$')

# Draw vector from (0,0) to vecA coordinates
plt.arrow( 0.0, 0.0, vecA[0], vecA[1], fc='red', ec='red', head_width=.1, head_length=.1, linewidth=2)

# Draw vector from vecA to vecB coordinates
plt.arrow(vecA[0], vecA[1], vecB[0], vecB[1], fc='blue', ec='blue', head_width=.1, head_length=.1, linewidth=2)

# Draw vector from (0,0) to vecA+vecB
vecAB = vecA + vecB
plt.arrow(0, 0, vecAB[0], vecAB[1], fc='black', ec='black', head_width=.1, head_length=.1, linewidth=2)
```

```python
# Line charts

x = [1,3,5,10]
y = [2,2.5,10,8]

# You can adjust size of the figure using figsize parameter
fig = plt.figure(figsize=(5,3))

plt.plot(x,y, 'o-')
plt.xlim(xmax=11, xmin=0) # Limits of the axes
plt.ylim(ymax=11, ymin=0)

plt.xlabel('X axis', fontsize=12) # Label text and its properties
plt.ylabel('Y axis', fontsize=12)

# Font properties for axes ticks
plt.tick_params(axis='both', which='major', labelsize=12)

# You can put a title and customize it using title function
plt.title('Figure title', fontsize=14, color='r')

plt.savefig('line-chart.png', bbox_inches='tight', dpi=300)
plt.savefig('line-chart.pdf', bbox_inches='tight', dpi=300)
```

```python
# Bar chart

x = [1,2,3,4,5]
y = [5,10,4,8,6]

plt.bar(x,y)
```

```python
# Multiple barchart

x = np.arange(5)
y1 = [5,10,4,8,6]
y2 = [6,9,5,9,5]

w = 0.3

plt.bar(x, y1, width=w, color='r')
plt.bar(x+0.4, y2, width=w)

plt.xticks(x + (w+0.4)/2, ('A', 'B', 'C', 'D', 'E'))

# Font properties for axes ticks
plt.tick_params(axis='both', which='major', labelsize=16)
```

```python
# Histograms can also be plotted similar to barcharts

# Let's create some number at random
x = list()
for i in range(100):
    #x.append(random.random())
    x.append(random.gauss(0,1))
    
plt.hist(x, alpha=0.5) # Hist function computes histogram

plt.xlabel('Random value: x', fontsize=12)
plt.ylabel('Histogram', fontsize=12)
```

```python
# Scatter plot
x, y = list(), list()

for n in range(100):
    #x.append(random.uniform(-5,5))
    #y.append(random.uniform(-5,5))
    x.append(random.gauss(0,2))
    y.append(random.gauss(2,1))
    
#plt.scatter(x,y, label='Blue')
plt.scatter(x,y, label='Blue', s=100, alpha=0.5, linewidth=0)

x2, y2 = list(), list()

for n in range(100):
    x2.append(random.uniform(-1,11))
    y2.append(random.uniform(-1,11))
    
#plt.scatter(x2, y2, c='r', label='Red')
plt.scatter(x2, y2, label='Red', s=100, c='r', alpha=0.5, linewidth=0)

#plt.legend(loc='upper left')
_ = plt.legend(loc='upper left', fontsize=14, scatterpoints=1)
```

```python
# Subplots are useful when you are planning to plot more than one figure at the same time.

fig, axarr = plt.subplots(2,2, figsize=(6,6))
plt.tight_layout()

xvals = [0,1,2,3,4]

axarr[0,0].plot(xvals, [2,4,6,8,10], 'o')
axarr[0,1].plot(xvals, [-2,-4,-6,-8,-10], 'r-o')
axarr[1,0].hist([1,1,1,2,3,4,4,5,5,5,5])
axarr[1,1].scatter([1,2,3,1.5,2.5], [3,1,2,2.5,4], s=50, marker='s')
```

```python
fig, axarr = plt.subplots(2, figsize=(6,6), sharex=True)
plt.tight_layout()

xvals = np.linspace(-2*np.pi, 2*np.pi, 100)
sinx = np.sin(xvals)
cosx = np.cos(xvals)

axarr[0].plot(xvals, sinx, c='r', linewidth=5)
axarr[0].set_ylabel('Sin(x)', fontsize=14)

axarr[1].plot(xvals, cosx, 'o-')
axarr[1].set_ylabel('Cos(x)', fontsize=14)
```

## Visualizing networks

Let's create a network described by adjacency matrix below.

$
A = \begin{bmatrix}
0 & 2 & 0 & 1\\ 
0 & 1 & 2 & 0\\ 
1  & 0 & 0 & 2\\ 
2 & 0 & 1 & 0
\end{bmatrix}
$

Nodes presented in columns and row are [A,B,C,D]

```python
import networkx as nx # import package

graph = nx.DiGraph() # Create a directed graph because out adjacency matrix is not symmetric.

# First create graph using add_edge function
graph.add_edge('A', 'B', weight=2)
graph.add_edge('A', 'D', weight=1)
graph.add_edge('B', 'B', weight=1)
graph.add_edge('B', 'C', weight=2)
graph.add_edge('C', 'A', weight=1)
graph.add_edge('C', 'D', weight=2)
graph.add_edge('D', 'A', weight=2)
graph.add_edge('D', 'C', weight=1)

print ('Number of nodes: {}'.format(graph.number_of_nodes()))
print ('Number of edges: {}\n'.format(graph.number_of_edges()))

# We can alternatively create graph using
A = np.zeros((4,4))
A[0,1] = 2
A[0,3] = 1
A[1,1] = 1
A[1,2] = 2
A[2,0] = 1
A[2,3] = 2
A[3,0] = 2
A[3,2] = 1

print(A)

graph2 = nx.from_numpy_matrix(A, create_using=nx.DiGraph())

print ('Number of nodes: {}'.format(graph2.number_of_nodes()))
print ('Number of edges: {}'.format(graph2.number_of_edges()))
```

```python
nx.draw_networkx(graph)

# Or you can export network as a file to visuzalize using Gephi or other programs.
nx.write_gexf(graph, 'network.gexf')
```
